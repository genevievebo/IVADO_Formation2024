Une classe peut Ãªtre considÃ©rÃ© de modÃ¨le pour crÃ©er des objets.  Les objets sont des instances de cette classe. Chaque objet peut avoir des propriÃ©tÃ©s (attributs) et des actions qu'il peut effectuer (fonctions). Une hiÃ©rarchie de classes est possible (classe parent, enfant).  Les classes enfant hÃ©ritent des mÃ©thodes et des attributs de leur parent et peuvent les redÃ©finir pour Ãªtre plus spÃ©cifique.

````python
class Animal:
    def __init__(self, nom, poids):
        self.nom = nom
        self.poids = poids
        self.unite_poids = "lbs"

    def manger(self):
        print(f"{self.nom} mange.")

    def communiquer(self):
        print(f"{self.nom} communique.")

class Chien(Animal):
    def communiquer(self):
        print(f"{self.nom} aboie.")

class Chat(Animal):
    def communiquer(self):
        print(f"{self.nom} miaule.")

class Poisson(Animal):
    pass
````

````python
chien = Chien("Pitou", 20)
chat = Chat("Minou", 5)
poisson = Poisson("Rouge", 0.1)

chien.manger()
chien.communiquer()

chat.manger()
chat.communiquer()

poisson.manger()
poisson.communiquer()

for animal in [chien, chat, poisson]:
    print(f"{animal.nom} pÃ¨se {animal.poids} {animal.unite_poids}")
````
